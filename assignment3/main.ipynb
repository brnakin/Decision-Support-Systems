```python
df_refined = df_refined.copy()

df_refined["goldDifference"] = (
    df_refined["blueTeamTotalGold"] - df_refined["redTeamTotalGold"]
)
df_refined["xpDifference"] = df_refined["blueTeamXp"] - df_refined["redTeamXp"]

df_refined.head()
```

```python
def drop_outliers(df):
    
    df_cleaned = df.copy()

    for column in df_cleaned.select_dtypes(include=["int64"]).columns:
        Q1 = df_cleaned[column].quantile(0.25)
        Q3 = df_cleaned[column].quantile(0.75)

        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_cleaned = df_cleaned[
            (df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)
        ]

    return df_cleaned
```

```python
df_refined.shape
```

```python
df_no_outliers = drop_outliers(df_refined)
```

```python
df_no_outliers.shape
```

```python
df_no_outliers.head(15)
```

```python
def min_max_scaling(df):

    df_scaled = df.copy()

    numeric_columns = df.select_dtypes(include=["int64"]).columns

    for column in numeric_columns:
        min_val = df[column].min()
        max_val = df[column].max()

        if min_val == max_val:
            df_scaled[column] = 0
        else:
            df_scaled[column] = (df[column] - min_val) / (max_val - min_val)

    return df_scaled
```

```python
df_scaled = min_max_scaling(df_no_outliers)
```

```python
df_scaled.head(15)
```

```python
df_scaled = df_scaled.drop(columns=["blueTeamTotalGold"])
df_scaled = df_scaled.drop(columns=["blueTeamXp"])
df_scaled = df_scaled.drop(columns=["redTeamTotalGold"])
df_scaled = df_scaled.drop(columns=["redTeamXp"])
```

```python
df_scaled.info()
```

```python
# Copy the DataFrame into new one to use for further homeworks

df_scaled_2 = df_scaled.copy()
```

```python
x = df_scaled.drop(columns=["blueWin"], axis=1)
y = df_scaled["blueWin"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Define parameter grid
param_grid = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100],
    "max_iter": [1000],
    "solver": ["lbfgs"],
}

# Initialize model
log_reg = LogisticRegression(random_state=42)
```

```python
# Perform Grid Search
print("Starting GridSearchCV...")
grid_search = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring="accuracy",
    verbose=1,
)
```

```python
# Fit the model
grid_search.fit(x_train, y_train)
```

```python
# Get best model
best_model = grid_search.best_estimator_
```

```python
# Make predictions
y_pred = best_model.predict(x_test)
```

```python
# Calculate feature importance (absolute values of coefficients)
feature_importance = pd.DataFrame(
    {"feature": x_train.columns, "importance": np.abs(best_model.coef_[0])}
).sort_values("importance", ascending=False)
```

```python
# Print results
print("\nBest Parameters:")
print(grid_search.best_params_)

print("\nModel Performance:")
print(f"Best Cross-Validation Score: {grid_search.best_score_:.4f}")
print(f"Test Set Accuracy: {accuracy_score(y_test, y_pred):.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
